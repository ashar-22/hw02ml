```python
import mlflow
import mlflow.xgboost
import pandas as pd
import numpy as np
import dagshub

dagshub.init(repo_owner='ashar-22', repo_name='hw02ml', mlflow=True)

test_identity = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_identity.csv')
test_transaction = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_transaction.csv')

test = pd.merge(test_transaction, test_identity, on='TransactionID', how='left')

submission = pd.DataFrame()
submission["TransactionID"] = test["TransactionID"]

run_id = "1cc69b76ea484f68bb45e1341a0e8e33"
model_uri = "runs:/" + run_id + "/fraud_pipeline"
loaded_model = mlflow.sklearn.load_model(model_uri)

preds = loaded_model.predict_proba(test)[:, 1]

submission["isFraud"] = preds
submission.to_csv("/kaggle/working/submission.csv", index=False)

print("âœ… Submission file created successfully!")
```
